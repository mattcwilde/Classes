```python
#define constants and coefficients
z = 2 # for He
A_n = 4 # number of nucleons
mu = (A_n * m_p) / z 
kid = 8.23e7 * (1 + z) / A_n # erg/g/K 
ke = 3.14e12 * (2 * m_p / mu)**(5/3)
kc = 2.23e12 * z**(2/3) * (2 * m_p / mu)**(4/3)

# make array of masses in log space
M = np.logspace(-5, 0) * M_sun

# Temperature given by Eric
T = 1e4 # K

a = (3 * M / (4 * np.pi))
A = kid * T * a
B = -kc * a**(4/3) - a * G * M
C = ke * a**(5/3)


def R_plus(M):
    rr = (np.sqrt(B**2 - 4*A*C) - B)/(2*A)
    return rr
def R_minus(M):
    rr = (-np.sqrt(B**2 - 4*A*C) - B)/(2*A)
    return rr
```

```python
%matplotlib inline
from __future__ import division
from astropy import constants as const
from astropy import units as u
import numpy as np
import matplotlib.pyplot as plt
# from numba import jit
plt.rc('font', family='serif', size=15)
```

```python
# universal constants
G = const.G.cgs.value
m_p = const.m_p.cgs.value
M_sun = const.M_sun.cgs.value
R_sun = const.R_sun.cgs.value
c = const.c.cgs.value
h = const.h.cgs.value
```

```python
FPS = np.array([6.22, 6.121, 0.006004, 0.16345, 6.50, 11.8440, 17.24, 1.065,
      6.54, 11.8421, -22.003, 1.5552, 9.3, 14.19, 23.73, -1.508, 1.79,
       15.13], dtype=float)
SLy = np.array([6.22, 6.121, 0.005925, 0.16326, 6.48, 11.4971, 19.105, 0.8928,
      6.54, 11.4950, -22.775, 1.5707, 4.3, 14.08, 27.80, -1.653, 1.50, 
       14.67], dtype=float)
d = {'FPS':FPS, 'SLy':SLy}
```

### Define functions

```python
def f0(x):
    """eq. 13"""
    f = 1 / (np.exp(x) + 1)
    return f


def EOS(rho,arg):
    """ eq. 14
        Takes in rho, model and gives P(rho)
    """
    xi = np.log10(rho)
    if not type(arg) == str:
        print "FPS or SLy not a string"
    zeta = ((d[arg][0] + d[arg][1]*xi + d[arg][2]*xi**3) / (1 + d[arg][3]*xi))*f0(d[arg][4]*(xi - d[arg][5]))
    zeta += (d[arg][6] + d[arg][7]*xi)*f0(d[arg][8]*(d[arg][9] - xi))
    zeta += (d[arg][10] + d[arg][11]*xi)*f0(d[arg][12]*(d[arg][13] - xi))
    zeta += (d[arg][14] + d[arg][15]*xi)*f0(d[arg][16]*(d[arg][17] - xi))
    return 10**zeta # P = EOS(rho, model)


def dPdrho(rho,P,arg):
    """eq. 19
    
    This function takes in (rho, P,'FPS' or 'SLy'). 
    It returns dP/drho = (P/rho) * dzeta/dxi
    """
    
    xi = np.log10(rho)
    z = ((d[arg][1] - d[arg][0]*d[arg][3]*xi**2 + 2*d[arg][2]*d[arg][3]*xi**3) / (1 + d[arg][3]*xi)**2
         -(d[arg][4]*(d[arg][0] + d[arg][1]*xi + 
                      d[arg][2]*xi**3) / (1 + 
                                          d[arg][3]*xi)*f0(d[arg][4]*(d[arg][5] - 
                                                                      xi))))*(f0(d[arg][4]*(xi - 
                                                                                            d[arg][5])))
    s = 0.
    for i in range(1,4):
        bracket2 = (d[arg][4*(i+1)-1] + d[arg][(4*(i+1)+1)-1]*(d[arg][(4*(i+1)-1)-1] + d[arg][(4*(i+1))-1] * xi)
                   *f0(d[arg][(4*(i+1)+1)-1]*(xi - d[arg][(4*(i+1)+2)-1])))
        s += f0(d[arg][(4*(i+1)+1)-1]*(d[arg][(4*(i+1)+2)-1] - xi))*bracket2
        
    dzdx = z + s
    return (P/rho)*dzdx # dP/drho


def dPdr(rho, P, r, M):
    """ TOV equation. 
        Takes in (rho, P, r, M)
    """
    bracket1 = rho + P/c**2
    bracket2 = M + 4*np.pi*(r**3)*(P/c**2)
    bracket3 = (1 - (2*G*M)/(r*(c**2)))**(-1)
    dPdr = (-G/r**2) * bracket1*bracket2*bracket3
    return dPdr


def drhodr(rho, P, r, M, arg):
    """ drho/dr
    """
    return dPdr(rho, P, r, M) / dPdrho(rho, P, arg)
    

def dM(rho,r,dr):
    """M(<r)"""
    
    dm = 4*np.pi*(r**2)*rho*dr
    return dm
```

```python
def NS_Profile(rho_c, arg):
    """ A function to step integrate the equations governing the NS stepping out in radius each step. 
    
    input: central density and model name.
    
    output: lists of each variable
    """
    # initialize lists
    rho_list = []
    P_list = []
    m_list = []
    r_list = []
    dPdrho_list = []

    # step size
    dr = 100.0 # cm
    
    # central starting values
    r = dr # cm
    rho = rho_c # density in g/cc
    m = dM(rho, r, dr) # mass in g
    P = EOS(rho, arg) # Pressure in dynes/cm2
    i = 0
    
    # add these values to list to keep track
    rho_list.append(rho)
    P_list.append(P)
    m_list.append(m)
    r_list.append(r)

    # while rho > rhoc*0.005:
    while rho > 0:
        # update density
        rho = drhodr(rho, P, r, m, arg)*dr + rho
        rho_list.append(rho)

        # update mass
        m = dM(rho, r, dr) + m
        m_list.append(m)

        # update pressure
        P = EOS(rho, arg)
        P_list.append(P)

        # update dP/drho
        dPdrho_list.append(dPdrho(rho, P, arg))

        # advance r
        r = r + dr
        r_list.append(r)

        # advance i
        i = i + 1

    return rho_list, P_list, m_list, r_list, dPdrho_list, i
```

```python
num_of_rhos = 50
# rho_c_list = np.logspace(np.log10(np.log10(2.5e14)), np.log10(np.log10(2e16)), num_of_rhos)
rho_c_list = np.linspace(2.5e14, 2e16, num_of_rhos)

FPS_max_m = []
FPS_max_rho = []
FPS_max_P = []
FPS_max_r = []
FPS_max_dPdrho = []

SLy_max_m = []
SLy_max_rho = []
SLy_max_P = []
SLy_max_r = []
SLy_max_dPdrho = []

for rho_i in rho_c_list:
    for model in d.keys():
        if model == 'FPS':
            rho_FPS, P_FPS, m_FPS, r_FPS, dPdrho_FPS, i = NS_Profile(rho_i, model)
            FPS_max_m.append(m_FPS[-2]) # -2 since last in list has rho < 0
            FPS_max_r.append(r_FPS[-2])
            FPS_max_rho.append(rho_FPS[-2])
            FPS_max_P.append(P_FPS[-2])
            FPS_max_dPdrho.append(dPdrho_FPS[-2])
            
        if model == 'SLy':
            rho_SLy, P_SLy, m_SLy, r_SLy, dPdrho_SLy, i = NS_Profile(rho_i, model)           
            SLy_max_m.append(m_SLy[-2])
            SLy_max_r.append(r_SLy[-2])
            SLy_max_rho.append(rho_SLy[-2])
            SLy_max_P.append(P_SLy[-2])
            SLy_max_dPdrho.append(dPdrho_SLy[-2])
            
```
